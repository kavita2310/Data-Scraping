```python
import requests
import pandas as pd
```

```python
from bs4 import BeautifulSoup as bs
```

```python
url='https://www.worldometers.info/coronavirus/'
```

```python
requests.get(url)
```

```python
text_extractor= requests.get(url).text
```

```python
#print(text_extractor)
```

```python
soup=bs(text_extractor,'html.parser')
```

```python
#print(soup)
```

```python
table_find=soup.findAll('table',{'id':'main_table_countries_yesterday'})
```

```python
#print(table_find)
```

```python
table_find=table_find[0].findAll('tr')
```

```python
#print(table_find)
```

```python
complete_table=[]
for i in table_find:
    td_data=i.findAll('td')
    temp_data=[]
    if 'china' in str(td_data):
        continue
        
    for i in range(1,len(td_data)):
        temp=td_data[i].text
        if ',' in str(temp):
            temp=temp.replace(',','')
        if '+' in str(temp):
            temp=temp.replace('+','')
        if '\n' in str(temp):
            temp=temp.replace('\n','')
        if 'N/A' in str(temp):
            temp=temp.replace('N/A','0')    
        temp_data.append(temp)
        
    complete_table.append(temp_data)    
    
```

```python
complete_table[8:]
```

```python
df=pd.DataFrame(complete_table[8:])
```

```python
df
```

```python
df.drop([14,15,16,17,18,19,20],axis=1,inplace=True)
```

```python
df.columns=(['Country','Other Total','Cases New','Cases Total','Deaths New','Deaths Total','Recovered New','Recovered Active','Cases Serious',
           'Critical Tot Cases','1M pop Deaths','1M pop Total','Tests Tests', '1M pop Population'])
```
